## Step 5 - Aggregation with GroupBy
* A groupby object with the "City Column", and aggregation function of sum and mean.
* A table that shows the gross income of each city, and determines the city with the highest total gross income.
* Optional - Use of appropriate methods & descriptions to explore other columns such as Unit Price, Quantity.

```python
df.groupby('City').agg(['sum', 'mean'])
```

```python
#A table that shows the gross income of each city 

# Get the sum of the durations per month
df.groupby('City')['gross income'].sum().reset_index()
```

```python
#determines the city with the highest total gross income.
df.groupby("City")['gross income'].sum().reset_index().max().reset_index()
```

```python
df.groupby('City').mean()
```

## Step 6 - Data Visualization
* Appropriate use of countplot to determine the branch with the highest sales record.

```python
#import seaborn as sns
sns.countplot(x='Branch', hue='City', data = df, palette = 'Set2').set_title('Branch with Sales', size=15);
```

```python
sns.countplot(x='Branch', data=df).set_title('Branches', size=15);
```

```python
#Plotting bar horizontal
sns.countplot(y='Branch', hue='City', data=df).set_title('Card Payment', size=15);
```

```python
#payment method & city
sns.countplot(x='Payment', hue='Payment', data=df, palette = "Set3").set_title('Card Payment', size=15);
```

```python
#payment method & city
sns.countplot(x='Payment', data=df, palette = "Set1").set_title('Card Payment', size=25);
```

```python
#highest & lowest sold product line.
sns.countplot(y='Product line', data=df, palette='Spectral_r', saturation = 0.9).set_title('Sale and its Product Lines', size=20)
```

* Highest product sale is Food and beverages
* Lowest product sale is Health and beauty

```python
#highest & lowest sold product line.
sns.countplot(y='Product line', hue='Payment', data=df, palette='Spectral_r', saturation = 0.9).set_title('Payment Method and its Product Lines', size=20)
```
